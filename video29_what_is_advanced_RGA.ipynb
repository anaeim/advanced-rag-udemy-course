# CRAG: Corrective RAG

[link](https://arxiv.org/pdf/2401.15884)


main idea of crag:<br>

This is an advanced RAG technique that is going to help us get more quality answers when performing retrieval. 

Augmentation generation and the basic concept of crag is pretty straightforward. 

We first want to start by taking our query, performing in the vector search, semantic search, and retrieving relevant documents from our vector store. After we have all those documents, we want to start in, self-reflect, to critique those documents, and to determine whether they are indeed relevant to our original query or not. <br>
If all of the documents are relevant to our query, then this is a happy flow and we simply want to augment our original prompt and send everything to the llm like we always do in Rag. <br> 
However, if you find document that we find out that are not relevant to our query, of course we want to filter them out, and we also want to perform some external search on the internet and get to more information. And to augment our prompt with that real time information we get from online. And after we have that, we can augment our prompt and send everything to the learn. And this technique is going to give us much more quality responsive.
